# ACCESSING DATAFRAMES

The data present in the DataFrames can be accessed using indexing and slicing.
INDEXING

**Indexing in DataFrames is of two types:**

Label based Indexing

Boolean Indexing.


Picture1.png

```python
#The loc() method is used for label based indexing. It accepts row/column labels as parameters.
#Dataframe.loc[start_row:end_row,start_col:end_col]

#To display the row with row label 2, the command is:
data.loc[2:6,'School ID':'Elementary, Middle, or High School']

#where the symbol : indicates all columns.
```

```python
data.iloc[2:6,0:1]  #when you are fetching using index values ( last value tha you mention is not inclusive)
```

```python
df.loc[2]
```

```python
# To access columns: use:
# DF.loc[:,start column:end column] Make sure not to miss the COLON BEFORE COMMA.
df.loc[:,"SAFETY_SCORE"]
```

**We can easily retrieve a slice of rows and columns using loc() and iloc() methods.**

The loc[] method is used to retrieve the group of rows and columns by labels or a boolean array present in the DataFrame. It takes only index labels, and if it exists in the called DataFrame, it returns the rows, columns, or DataFrame.

**The .iloc[] method is used retrieve the group of rows and columns at particular positions in the index so it only takes integers. It is used when the index label of the DataFrame **is other than numeric series of 0,1,2,	,n, or in the case when the user
does not know the index label

```python
data.iloc[::10,::2]

#every 10th record and every alternate column
```
